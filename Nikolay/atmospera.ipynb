```python
def  work_with_end_date(x):
    x = pd.to_datetime(x)
    if x.minute >0:
        return x.replace(microsecond=0, second=0, minute=0) + timedelta(hours=1)
    else:
        return x.replace(microsecond=0, second=0, minute=0)
```

```python
atmos['clean_end_hour'] = atmos['end_ts'].apply(lambda x: work_with_end_date(x))
```

```python
atmos['start_date']= atmos['start_date'].astype('datetime64')
atmos['end_date']= atmos['end_date'].astype('datetime64')
```

```python
def sum_hour(x):
    start_date, hours = x
    new_date = start_date  + timedelta(hours=hours.hour)
    return new_date
```

```python
atmos['start_data_with_hour'] =atmos[['start_date','clean_strat_hour']].apply(sum_hour, axis=1)
```

```python
atmos['end_data_with_hour'] =atmos[['end_date','clean_end_hour']].apply(sum_hour, axis=1)
```
