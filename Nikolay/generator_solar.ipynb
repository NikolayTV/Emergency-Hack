```python
import pandas as pd
import numpy as np
import pvlib as pv
import datetime
```

```python
geo_data=pd.read_csv('../data/geo_data.csv')
```

```python
sample_nine = geo_data[geo_data['road_id']== 9]
#sample_onefour = geo_data[geo_data['road_id']== 14]
```

```python
result = pd.DataFrame (columns = ['lon', 'lat', 'range_km', 'road_id'])
for frame in [sample_nine]:#, sample_onefour]:
    road_id = frame['road_id'].iloc[0]
    for i in np.arange(0,max(frame['road_km']), 50 ):
        df =frame[(i < frame['road_km']) & (frame['road_km'] < i+50)]
        lon = df['lon_geoc'].mean()
        lat = df['lat_geoc'].mean()
        range_km = i
        result = result.append({'lon': lon, 'lat': lat, 'range_km': range_km, 'road_id': road_id}, ignore_index=True)
        result['range_km']= result['range_km'].astype(int)
        result['road_id']= result['road_id'].astype(int)
```

# Mapper to frames

```python
def get_solar(x):
    datetime, lon, lat = x.values
    df = pv.solarposition.get_solarposition(time = datetime, latitude= lat, longitude = lon)
    df['lon'] = lon
    df['lat'] = lat
    df.index.name = 'datetime'
    df.reset_index(inplace=True)
    return df
     
```

```python
import time
start = time.process_time()
#enriched = pd.DataFrame(columns = ['apparent_zenith',  'zenith', 'apparent_elevation', 'elevation', 'azimuth', 'equation_of_time'])
enriched = pd.DataFrame()
enriched = merged[['datetime', 'lon', 'lat']].dropna().apply(get_solar, axis=1)
print(time.process_time() - start)
```
